### Labeling input and output columns

```python
X= df.drop("selling_price", axis=1)
Y=df["selling_price"]
```

```python
# splitting data into train and test data
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.2, random_state=2)
```

### Using Linear Regression

```python
lr= LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
print ("r2_score for linear regression is :", r2_score(y_test, y_pred))
```

### Using Lasso Regression

```python
lasso= Lasso(alpha=1, max_iter=1000)
lasso.fit(X_train, y_train)
```

```python
y_pred2= lasso.predict(X_test)
```

```python
print ("r2_score for lasso regression is :", r2_score(y_test, y_pred2))
```

### Using RandomForestRegression

```python
rfr= RandomForestRegressor()
rfr.fit(X_train, y_train)
```

```python
y_pred3= rfr.predict(X_test)
```

```python
print ("r2_score for RandomForestRegression is :", r2_score(y_test, y_pred3))
```
